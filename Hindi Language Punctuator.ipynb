```python
s = ''

for n in output_json:
  result = n['word'].replace('▁',' ') + n['entity'].replace('0','')
  s += result
```

```python
s=s.replace('.','|')
```

```python
from spacy import displacy
```

```python
empty_ner = []
for n in output_json:
  if n['entity'] != '0':
    if n['entity']=='.': 
      n['entity']='|'
    temp = [n['start'],n['end'],n['entity']]
    empty_ner.append(temp)
```

```python
empty_ner
```

```python
colors = {"0": "linear-gradient(90deg, #999999, #cccccc)",
          ",": "linear-gradient(90deg, #aa9cfc, #fc9ce7)",
          "|": "linear-gradient(90deg, #aa9cfc, #fc9ce7)",
          "?": "linear-gradient(90deg, #aa9cfc, #fc9ce7)",
          ":": "linear-gradient(90deg, #aa9cfc, #fc9ce7)",
          "-": "linear-gradient(90deg, #aa9cfc, #fc9ce7)"
          }
options = {"ents": ["0", ",","|","?",":",'-'], "colors": colors}
```

```python
[{"start": x[0], "end": x[1], "label": x[2]} for x in empty_ner]
```

```python
ex = [{"text": text,
       "ents": [{"start": x[0], "end": x[1], "label": x[2]} for x in empty_ner],
       "title": "Deep Learning powered Punctuation Corrector"}]
```

```python
html = displacy.render(ex, style="ent", manual=True, options=options)
```
